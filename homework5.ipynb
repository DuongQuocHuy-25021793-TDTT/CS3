```python
#W6A16:
n , m = map(int,input().split())
a = []
for i in range(n):
    row = list(map(int,input().split()))
    a.append(row)
for i in range(n):
    for j in range(m):
        print(f"{a[i][j]:>4}",end=' ') # căn phải 4 vị trí , >4 → căn phải , <4 → căn trái , ^4 → căn giữa
    print()
```

```python
#W6A17:
n = int(input())
a = []
for i in range(n):
    b = list(map(int,input().split()))
    a.append(b)
# đường chéo chính
for i in range(n):
    for j in range(n):
        if i == j :
            print(a[i][j])
    print()
#đường chéo phụ
for i in range(n):
    for j in range(n):
        if i + j == n-1:
            print(a[i][j])
    print()
```

```python
#W6A21:
s = input().split()

d = {}

for i, word in enumerate(s):
    if word not in d:
        d[word] = []
    d[word].append(i)

for key in d:
    d[key] = tuple(d[key])

print(d)
```

```python
#W6A26:
# phần tử cực đại địa phương là phần tử lớn hơn 2 phần tử kề nó
n = int(input())
a = list(map(int,input().split()))
tong = 0
for i in range(1,n-1):
    if a[i-1] < a[i] and a[i] > a[i+1]:
        tong += a[i]
print(tong)
```

```python
#W6A28:
n = int(input())
a = list(map(int,input().split()))
tang = True
giam = True
for i in range(1,n):
    if a[i] <= a[i-1] :
        tang = False
    else :
        giam = False
if tang :
    print('Dãy Tăng')
elif giam :
    print('Dãy Giảm')
else:
    print("Vô Danh")
```
