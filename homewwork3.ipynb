```python
#bai4:
h = int(input("Nhập chiều cao h: "))

for i in range(1, h + 1):
    # In khoảng trống phía trước
    print(" " * (h - i), end="")

    if i == 1:
        # Đỉnh tam giác
        print("*")
    elif i == h:
        # Đáy tam giác: in toàn dấu *
        print("*" * (2 * h - 1))
    else:
        # Các dòng ở giữa: * ... *
        print("*" + " " * (2 * i - 3) + "*")
```

```python
#bai8:
def gcd(a,b):
    while b!=0:
        a , b = b , a%b
    return a
if __name__ == "__main__":
    a,b = map(int, input().split())
    print(gcd(a,b))
```

```python
#bai10:
from math import*
a = int(input())
tong = 1
for i in range(2,isqrt(a)+1):
    if a % i == 0:
        tong += i
        j = a // i
        if j != i:
            tong += j
print(tong)
```

```python
#bai14:
max_value = None # nghĩa là chưa có số nào
min_value = None 
while True:
    x = int(input())
    if x == -1:
        break
    if max_value is None : # nếu là số duy nhất
        max_value = x
        min_value = x
    else:
        if x > max_value :
             max_value = x
        if x < min_value:
             min_value = x
print('so lon nhat :',max_value)
print('so nho nhat :',min_value)

```

```python
#bai19:
A = int(input())
fn1 , fn2 = 1 , 1
max_fib = 1 # lưu số fibo không vượt quá A
while fn2 <= A:
    max_fib = fn2
    fn1 , fn2 = fn2 , fn1 + fn2 # cập nhật 2 số fibo tiếp theo
print(max_fib)
```

```python
#bai20:
from math import*
def nt(n):
    if n <= 1:
        return False
    for i in range(2,isqrt(n)+1):
        if n % i == 0:
            return False
    return True
if __name__ == "__main__":
    n = int(input())
    dem = 0
    while n !=0:
        x = n % 10
        if nt(x):
            dem +=1 
        n //=10
    print(dem)
```
