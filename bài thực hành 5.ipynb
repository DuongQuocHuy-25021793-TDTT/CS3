```python
#5WA3:
from math import*
def nt(n):
    if n <= 1:
        return False
    for i in range(2, isqrt(n)+1):
        if n % i == 0:
            return False
    return True
if __name__ == "__main__":
    n = int(input())
    print(nt(n))
```

```python
#5WA4:
from math import*
def so_hoan_hao(n):
    tong = 1
    if n <= 1:
        return False
    for i in range(2,isqrt(n)+1):
        if n % i == 0:
            tong += i
            j = n / i
            if j != i :
                tong += j 
    return tong
if __name__ == "__main__":
    n = int(input())
    if so_hoan_hao(n) == n:
        print('true')
    else:
        print('false')
```

```python
#5WA7:
def phep_toan(a ,b , phep):
    if phep == '+':
        return a + b
    if phep == '-':
        return a - b
    if phep == '*':
        return a * b
    if phep == '/':
        if b == 0 :
            return False
        else:
            return a / b
if __name__ == "__main__":
    a , b = map(int , input().split())
    phep = str(input())
    print(phep_toan(a ,b , phep))    



```
