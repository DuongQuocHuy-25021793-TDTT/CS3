```python
#5WA4:
from math import*
def so_hoan_hao(n):
    tong = 1
    if n <= 1:
        return False
    for i in range(2,isqrt(n)+1):
        if n % i == 0:
            tong += i
            j = n / i
            if j != i :
                tong += j 
    return tong
if __name__ == "__main__":
    n = int(input())
    if so_hoan_hao(n) == n:
        print('true')
    else:
        print('false')
```

```python
#5WA7:
def phep_toan(a ,b , phep):
    if phep == '+':
        return a + b
    if phep == '-':
        return a - b
    if phep == '*':
        return a * b
    if phep == '/':
        if b == 0 :
            return False
        else:
            return a / b
if __name__ == "__main__":
    a , b = map(int , input().split())
    phep = str(input())
    print(phep_toan(a ,b , phep))    



```

```python
#5WA8:
def hamming_distance(a, b):
    # Đổi hai số sang nhị phân, bỏ '0b'
    bin_a = bin(a)[2:]
    bin_b = bin(b)[2:]
    # Thêm số 0 vào đầu để hai chuỗi bằng độ dài
    max_len = max(len(bin_a), len(bin_b))
    bin_a = bin_a.zfill(max_len)
    bin_b = bin_b.zfill(max_len)
    # Đếm số vị trí khác nhau
    count = 0
    for i in range(max_len):
        if bin_a[i] != bin_b[i]:
            count += 1
    return count
print(hamming_distance(5, 10))  
```

```python
#5WA9:
def tong_uoc(n):
    x = 0
    tong = 0
    while n != 0 :
        x = n % 10
        tong += x
        n //= 10
    return tong
if __name__ == "__main__":
    n =  int(input())
    print(tong_uoc(n))    
```

```python
#bai10:
def is_isomorphic(s, t):
    if len(s) != len(t):
        return False

    map_st = {}   # ánh xạ ký tự từ s → t
    map_ts = {}   # ánh xạ ký tự từ t → s

    for c1, c2 in zip(s, t):
        # Nếu c1 đã có ánh xạ, phải trùng với c2
        if c1 in map_st:
            if map_st[c1] != c2:
                return False
        else:
            map_st[c1] = c2

        # Ngược lại: nếu c2 đã ánh xạ, phải về c1
        if c2 in map_ts:
            if map_ts[c2] != c1:
                return False
        else:
            map_ts[c2] = c1

    return True
print(is_isomorphic("abca", "zbxz"))  
print(is_isomorphic("abb", "xyz"))    
```
