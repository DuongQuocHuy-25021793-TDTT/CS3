```python
#W6A9:
# Nhập hai dictionary
d1 = eval(input())
d2 = eval(input())

# Gộp dictionary
result = {}

for key in set(d1.keys()) | set(d2.keys()):
    result[key] = d1.get(key, 0) + d2.get(key, 0)

# Sắp xếp theo key tăng dần
result = dict(sorted(result.items()))

# In kết quả
print(result)
```

```python
#W6A10:

# Nhập dữ liệu
lst = list(map(int, input().split()))
k = int(input())

pairs = set()

n = len(lst)
for i in range(n):
    for j in range(i + 1, n):
        if lst[i] + lst[j] == k:
            a, b = sorted((lst[i], lst[j]))
            pairs.add((a, b))

# Chuyển sang list và sắp xếp
result = sorted(pairs, key=lambda x: x[0])

# In kết quả
print(result)
```

```python
#W6A12:
a = list(map(int,input().split()))
max = 0
cnt = a[0]
for x in a:
    dem = a.count(x)
    if dem > max or (dem == max and x < cnt ) :
        dem = max
        cnt = x
print(cnt)
```
